```python
# 3 x 128 x 128
# 3x16x16(=768)x8x8(=64)
# 64个token，每个token长384

from PIL import Image
import torchvision.transforms as transforms

# 加载图片
image = Image.open('./image/user3_3.jpg')

# 定义变换，将图片resize为128x128
transform = transforms.Compose([
    transforms.Resize((128, 128)),
    transforms.ToTensor()
])
# 对图片应用变换
resized_image = transform(image)
resized_image.numpy()

# 打印调整后的图片尺寸
print("Resized image size:", resized_image.size())
```

```python
import torch
import torch.nn as nn

in_c=3
embed_dim=384
img_size=128
patch_size=16

processed_tensors = []
for i in range(1,379):
    # Read the image
    image = Image.open("./facesFromFrames_new/{}.jpg".format(i))
    image = image.resize((img_size, img_size))

    transform = transforms.ToTensor()
    image_tensor = transform(image)

    # Reshape the image tensor to 1x3x128x128
    image_tensor = image_tensor.unsqueeze(0)

    # Process the image tensor using the provided code
    data = nn.Conv2d(in_c, embed_dim, kernel_size=patch_size, stride=patch_size)
    output = data(image_tensor)
    print("123",output.size())

    # Reshape the output tensor to 768 x 14 x 14
    output = output.view(embed_dim, img_size//patch_size, img_size//patch_size)

    # Multiply the last two dimensions together
    output = output.view(embed_dim, -1)
    # Transpose the tensor to obtain a shape of (64, 384)
    output = output.transpose(0, 1)

    print(output.size())
    processed_tensors.append(output)

final_tensor = torch.cat(processed_tensors, dim=0)
print(final_tensor.shape)
```

```python
from PIL import Image
import numpy as np
# 创建一个空的列表，用于存储切割后的图片
split_images = []

# 遍历图片的宽度和高度
for i in range(0, 128, 16):
    for j in range(0, 128, 16):
        # 切割图片
        split_image = resized_image[:, i:i+16, j:j+16]
        # 将切割后的图片添加到列表中
        split_images.append(split_image.numpy().astype('uint8'))

# 打印切割后的图片数量
print("Number of split images:", len(split_images))

# 遍历存储切割后图片的列表
for i, img in enumerate(split_images):
    # 将图片从3x16x16转换为16x16x3
    img = np.transpose(img, (1, 2, 0))
    # 将numpy数组转换为图像
    pil_img = Image.fromarray(img)
    # 将图像保存为文件
    pil_img.save(f'./image/image_{i}.png')
```
